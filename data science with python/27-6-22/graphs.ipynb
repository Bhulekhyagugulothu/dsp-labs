## pie chart


```python
## explode separates the part of chart 
import matplotlib.pyplot as plt
import numpy as np
y=np.array([30,40,25,5])
plt.pie(y,explode=(0,0.1,0,0))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([30,40,25,5])
co=["r","b","yellow","pink"]
plt.pie(y,colors=co,labels=["dsp","cd","wt","or"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([30,40,25,5])
co=["r","b","yellow","pink"]
plt.pie(y,colors=co,labels=["dsp","cd","wt","or"])
plt.legend(title="subjects")
plt.show()
```

## bar graph

```python
import matplotlib.pyplot as plt
Country=["usa","india","germany","uk","france"]
GDP_per_captia=[48000,38000,50000,45000,44000]
plt.bar(Country,GDP_per_captia,color="pink")
plt.title("COUNTRY VS GDP_PER_CAPTIA ($)")
plt.xlabel("Country")
plt.ylabel("GDP_PER_CAPTIA($)")
plt.show()
```

```python
import matplotlib.pyplot as plt
Country=["usa","india","germany","uk","france"]
GDP_per_captia=[48000,38000,50000,45000,44000]
plt.barh(Country,GDP_per_captia,color="blue")
plt.title("COUNTRY VS GDP_PER_CAPTIA ($)")
plt.xlabel("Country")
plt.ylabel("GDP_PER_CAPTIA($)")
plt.show()
```

# histogram

```python
import matplotlib.pyplot as plt
num=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(num)
plt.xlabel("NUMBERS")
plt.ylabel("frequency")
plt.title("NUM  VS  FQY")
plt.show()
```

```python
import matplotlib.pyplot as plt
num=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(num,bins=5,color="pink")
plt.xlabel("NUMBERS")
plt.ylabel("frequency")
plt.title("NUM  VS  FQY")
plt.show()
```

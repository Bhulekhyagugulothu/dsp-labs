```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,20])
ypoints=np.array([0,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([5,20],[5,200],"--")
plt.show()
```

# scatter

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,4,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,4,1,10,5,7])
plt.plot(ypoints,marker="o",mec="r",color="yellow")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,4,1,10,5,7])
plt.plot(ypoints,marker="*",mec="blue",color="yellow")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,4,1,10,5,7])
plt.plot(ypoints,marker="*",mec="blue",color="r",linewidth="10")
plt.show()
```
